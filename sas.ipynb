```python
import random

class Bank:
    def _init_(self, name, balance):
        self.name = name
        self.balance = balance
        # Generates a random 16-digit number as a string
        self.account_number = ''.join([str(random.randint(0, 9)) for _ in range(16)])

    def deposit(self, amount):
        self.balance += amount
        print(f"Successfully deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Successfully withdrew ${amount}. Remaining balance: ${self.balance}")
        else:
            print("Insufficient funds!")

    def show_details(self):
        print(f"\n--- Account Details ---")
        print(f"Name: {self.name}")
        print(f"Account Number: {self.account_number}")
        print(f"Balance: ${self.balance}")
        print("-----------------------")

```
