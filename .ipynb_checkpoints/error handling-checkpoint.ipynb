```python
try:
    a = int(input('Enter a number: '))
    b = int(input('Enter a number: '))
    c = a / b
    print(round(c,2))
except ZeroDivisionError:
    print(f'Error: Numbers cannot be divided by 0')

except ValueError:
    print(f'Error: Input field only accept number integer values')
```

```python
file_name = input('Enter your file name: ')
extension = input('Enter your file extension: ')

try:
    with open(f'{file_name}.{extension}', 'r') as file:
        content= file.read('')
        
except FileNotFoundError:
    print(f'Error: {file_name}.{extension} not found in directory.')

else:
    print(content)

finally:
    print('Program ended!')
```

```python
class AgeExceptionError(Exception):
    pass
```

```python
def checkAge(age):
    if age>= 18:
        print(f'You are eligible to vote!')

    else:
        raise AgeExceptionError(f'Error: {age} years are not eligible to vote!')
```

```python
try:
    age = int(input('Enter your age: '))
    checkAge(age)

except AgeExceptionError as e:
    print(e)

except ValueError:
    print(f'\nError: Age value must be an integer.')
```

```python
class InsufficientFundError(Exception):
    pass
```

```python
def checkFunds(amount):
    if amount <= balance:
        print('Enter your PIN number.')

    else:
        raise InsufficientFundError(f'Error: {amount} must be less than the balance.')
```

```python
try:
    balance = 50000
    amount = int(input('Enter your withdraw amount: '))
    checkFunds(amount)
except InsufficientFundError as e:
    print(e)
except ValueError:
    print(f'\nError: Age value must be an integer.')
```

```python
class InvalidTemperatureError(Exception):
    pass
```

```python
def checkTemperature(temp):
    if temp < -273.15:
        print(f'Temperature below absolute zero is not possible.')
    else:
        raise InvalidTemperatureError(f'Error: {temp} must be greater than absolute temperature.')
```
