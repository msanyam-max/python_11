```python
'''
Python Operators
-------------------
1. Arithmetic Operator
    -> +, -, *, /, % (modulus / Reminder), // (float division)
2. Logical Operator
    -> And, Or, Not
3. Comparision Operator
    -> =, !=, >, <, >=, <=
'''
```

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
rem = num1 % num2
fd = num1 // num2
```

```python
print(add)
print(sub)
print(mul)
print(div)
print(rem)
print(fd)
```

```python
num3 = 10
```

```python
(num1 > num2) != (num2 < num3)
```
