# Constructor

```python
class Student:
    def __init__(self, name, section, no_of_subjects, past_failures, current_grade):  #constructor
        self.name = name
        self.section = section
        self.no_of_subjects = no_of_subjects                       # these are instance variables.
        self.past_failures = past_failures
        self.current_grade = current_grade
        self.school_name = 'ABC School'
        self.school_location = 'Kumaripati, Lalitpur'
        
    def student_details(self):
        print(f'School Name: {self.school_name}')
        print(f'School Location: {self.school_location}')
        print(f'Name of student :{self.name}')
        print(f'Section of {self.name} : {self.section}')
        print(f'Number of subjects : {self.no_of_subjects}')
        print(f'Past failures : {self.past_failures}')
        print(f'Currect grades: {self.current_grade}')
```

```python
s = Student('Bob', 'C11', 4, 2, 'A')
```

```python
s.student_details()
```

```python
class SmartCalculator:
    def add(self, a, b):
        num = a+b
        return num
     
    def is_even(self, num):
        if (num % 2 == 0):
            return 'even'

        else:
            return 'odd'
```

```python
s = SmartCalculator()
```

```python
s.add(4,6)
```

```python
s.is_even(55)
```

```python
# ALTERNATIVELY
class SmartCalculator:
    def __init__(self, num1, num2, num3):
        self.num1 = num1
        self.num2 = num2
        self.num3 = num3

    def add(self):
        return f'Sum of {self.num1} and {self.num2} is: {self.num1 + self.num2}'

    def is_even(self):
        if self.num3 % 2 == 0:
            return 'Even' 
        else:
            return 'Odd'
```

```python
# JOB APPLICATION

class Developer:
    def __init__(self, name, primary_language, years_of_service):
        self.name = name
        self.primary_language = primary_language
        self.years_of_service = years_of_service
        self.role = 'Software Engineer'
        self.environment = 'Remote'
        
    def coding_status(self):
        return f'{self.name} is currently coding in {self.primary_language}'

    def experience(self):   
        print(f'Role of the employee: {self.role}')    
        if self.years_of_service >= 5:
            return 'Senior'
        else:
            return 'Junior'
```

```python
d = Developer('John', 'Python', 6)
```

```python
d.experience()
```

# Inheritance

```python
#parent class
class School:
    def __init__(self, school_name, school_location, school_contact):
        self.school_name = school_name
        self.school_location = school_location
        self.school_contact = school_contact

    def school_details(self):
        print(f'School Name: {self.school_name}')
        print(f'School Location: {self.school_location}')
        print(f'School contact: {self.school_contact}')
```

```python
#child class
class Student(School):
    def __init__(self, student_name, student_section, past_failures, student_current_grade, school_name, school_location, school_contact):  #constructor
        super().__init__(school_name, school_location, school_contact)
        self.student_name = student_name
        self.student_section = student_section                      # these are instance variables.
        self.student_past_failures = past_failures
        self.student_current_grade = student_current_grade
        
    def student_details(self):
        print(f'Name of student :{self.student_name}')
        print(f'Section of {self.student_name} : {self.student_section}')
        print(f'Past failures : {self.student_past_failures}')
        print(f'Currect grades: {self.student_current_grade}')
```

```python
s = Student('charles', 'C7',3, 'B', 'ABC School', 'Kathmandu', '02002020')
```

```python
s.student_details()
s.school_details()
```

```python
## multiple inheritance
```

```python
class Teacher(School):
    def __init__(self, teacher_name, teacher_contact, total_classes, teacher_qualification, 
                 school_name, school_location, school_contact):
        super().__init__(school_name, school_location, school_contact)
        self.teacher_name = teacher_name
        self.teacher_contact = teacher_contact
        self.total_classes = total_classes
        self.teacher_qualification = teacher_qualification

    def teacher_details(self):
        print(f'teacher name: {self.teacher_name}')
        print(f'teacher contact: {self.teacher_contact}')
        print(f'total classes: {self.total_classes}')
        print(f'teacher qualification: {self.teacher_qualification}')
```

```python
t = Teacher('Bob', '012484844', 4, 'Bachelors', 'XYZ School', 'Bhaktapur', '19393999944')
t.teacher_details()
t.school_details()
```

```python
#parent class
class Order:
    def __init__(self, order_goods, order_quantity, order_number):
        self.order_goods = order_goods
        self.order_quantity = order_quantity
        self.order_number = order_number

    def order_details(self):
        print(f'Order Goods: {self.order_goods}')
        print(f'Order Quantity: {self.order_quantity}')
        print(f'Order Number: {self.order_number}')
```

```python
#child class
class Customer(Order):
    def __init__(self, customer_name, customer_contact, customer_id, customer_address, order_goods, order_quantity, order_number):
        super().__init__(order_goods, order_quantity, order_number)
        self.customer_name = customer_name
        self.customer_contact = customer_contact
        self.customer_id = customer_id
        self.customer_address = customer_address

    def customer_details(self):
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Contact: {self.customer_contact}')
        print(f'Customer ID: {self.customer_id}')
        print(f'Customer Address: {self.customer_address}')
```

```python
c = Customer('Sam', '01-333488', '9333413', 'Pokhara', 'Meat', 500, 19292923303)
c.customer_details()
c.order_details()
```

```python
#child class
class Staff(Order):
    def __init__(self, staff_name, staff_contact, staff_id, order_goods, order_quantity, order_number):
        super().__init__(order_goods, order_quantity, order_number)
        self.staff_name = staff_name
        self.staff_contact = staff_contact
        self.staff_id = staff_id

    def staff_details(self):
        print(f'Staff Name: {self.staff_name}')
        print(f'Staff Contact: {self.staff_contact}')
        print(f'Staff ID: {self.staff_id}')
```

```python
s = Staff('Sameul', '01-3367488', '9823323413', 'Meat', 500, 19292923303)
s.staff_details()
s.order_details()
```

```python
# Parent class
class School:
    def __init__(self, school_name, school_location, school_contact):
        self.school_name = school_name
        self.school_location = school_location
        self.school_contact = school_contact

    def school_details(self):
        print(f'School Name: {self.school_name}')
        print(f'School Location: {self.school_location}')
        print(f'School contact: {self.school_contact}')
```

```python
class Teacher(School):
    def __init__(self, teacher_name, teacher_contact, total_classes, teacher_qualification,
                 school_name, school_location, school_contact):
        super().__init__(school_name, school_location, school_contact)
        self.teacher_name = teacher_name
        self.teacher_contact = teacher_contact
        self.total_classes = total_classes
        self.teacher_qualification = teacher_qualification

    def teacher_details(self):
        print(f'teacher name: {self.teacher_name}')
        print(f'teacher contact: {self.teacher_contact}')
        print(f'total classes: {self.total_classes}')
        print(f'teacher qualification: {self.teacher_qualification}')
```

```python
#child class
class Student(Teacher):
    def __init__(self, student_name, student_section, student_past_failures, student_current_grade,  
                 school_name, school_location, school_contact, 
                 teacher_name, teacher_contact, total_classes, teacher_qualification):  #constructor
        super().__init__(school_name, school_location, school_contact, teacher_name, teacher_contact, total_classes, teacher_qualification)
        self.student_name = student_name
        self.student_section = student_section                      # these are instance variables.
        self.student_past_failures = student_past_failures
        self.student_current_grade = student_current_grade
        
    def student_details(self):
        print(f'Name of student :{self.student_name}')
        print(f'Section of {self.student_name} : {self.student_section}')
        print(f'Past failures : {self.student_past_failures}')
        print(f'Currect grades: {self.student_current_grade}')
```

```python
s = Student('John', 'A13', 5, 'A', 'Bob', 1234567890, 5, 'Bachelors', 'ABC', 'Kumaripati', 9876543210)
```

```python
s.student_details()
```

```python
s.teacher_details()
```

```python
s.school_details()
```

## Encapsulation

```python
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone, customer_email):
        self.__customer_id = customer_id
        self.__customer_name = customer_name
        self.__customer_address = customer_address
        self.__customer_phone = customer_phone
        self.__customer_email = customer_email

    def customer_details(self):
        print(f'Customer ID: {self.__customer_id}')
        print(f'Customer Name: {self.__customer_name}')
        print(f'Customer Address: {self.__customer_address}')
        print(f'Customer Phone: {self.__customer_phone}')
        print(f'Customer Email: {self.__customer_email}')
```

```python
class Order(Customer):
    def __init__(self, order_id, order_date, order_time, order_destination, customer_id, customer_name, customer_address, customer_phone, customer_email):
        super().__init__(customer_id, customer_name, customer_address, customer_phone, customer_email)
        self.__order_id = order_id
        self.__order_date = order_date
        self.__order_time = order_time
        self.__order_destination = order_destination

    def order_details(self):
        print(f'Order Id: {self.__order_id}')
        print(f'Order Date: {self.__order_date}')
        print(f'Order Time: {self.__order_time}')
        print(f'Order Destination: {self.__order_destination}')
```

```python
o = Order(1, '20th May 2026', '12:30 PM', 'Bhaktapur', 12, 'Ram', 'Lalitpur', '9876543456', 'ram@gmail.com')
```

```python
o.order_details()
o.customer_details()
```

```python
class Brand:
    def __init__(self, brand_id, brand_name):
        self._brand_id = brand_id
        self._brand_name = brand_name

    def brand_details(self):
        print(f'Brand ID: {self._brand_id}')
        print(f'Brand Name: {self._brand_name}')
```

```python
class Product(Brand):
    def __init__(self, product_id, product_name, product_desc, product_price, brand_id, brand_name):
        super().__init__(brand_id, brand_name)
        self._product_id = product_id
        self._product_name = product_name
        self._product_desc = product_desc
        self._product_price = product_price

p._    def product_details(self):
        print(f'Product ID: {self._product_id}')
        print(f'Product Name: {self._product_name}')
        print(f'Product Descrioption: {self._product_desc}')
        print(f'Product Price: Rs.{self._product_price}')
```

```python
p = Product(1, 'Nokia Mobile', 'Keypad Mobile', 3000, 123, 'Nokia')
```

```python
p.product_details()
p.brand_details()
```

```python
class Facebook:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'Facebook:\n{send_by} sent you a message.')
        print(f'Message: {message}')
           
class WhatsApp:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'WhatsApp:\n{send_by} sent you a message.')
        print(f'Message: {message}')
    
class Viber:
    def send_message(self, send_by, send_to, message):
        print(f'Profile Name: {send_to}')
        print(f'Viber:\n{send_by} sent you a message.')
        print(f'Message: {message}')
```

```python
def message(message_object, send_by, send_to, message):
    message_object.send_message(send_by, send_to, message)
```

```python
message(Viber(),'Ram', 'Shyam', 'Hello!!')
```

```python
## Abstraction
```

```python
from  abc import ABC, abstractmethod
```

```python
class MobileBanking(ABC):
    @abstractmethod
    def statement(self):
        pass

    @abstractmethod
    def user_account_details(self):
        pass

    def connection(self):
        return 'Mobile Banking Server connected to Database.'

    def username(self, uname):
        return f'Username: {uname}'
```

```python
class ShowDetails(MobileBanking):
    def show_statement(self):
        return f'Choose Statement.'

    def statement(self):
        return 'Statement from date1 to date2'
        
    def user_account_details(self):
        return 'User Account'
```

```python
s = ShowDetails()
```

```python
s.statement()
```

```python
print(s.connection())
print(s.username('Bob'))
```
