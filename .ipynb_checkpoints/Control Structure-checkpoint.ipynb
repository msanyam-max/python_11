```python
temperature = float(input('Enter temperature value: '))

if temperature >= 25: # False
    print('Its hot today.')
else:
    print('Its cold today.')
```

```python
temperature = float(input('Enter temperature value: '))

if temperature >= 30: # False
    print('Its hot today!!')
elif temperature >= 20 and temperature < 30: # False   True   False
    print('Its warm today!!')
else:
    print('Its cold today!!')
```

### Student Grade Calculator

```python
sub1 = float(input('Enter marks of subject 1: '))
sub2 = float(input('Enter marks of subject 2: '))
sub3 = float(input('Enter marks of subject 3: '))
sub4 = float(input('Enter marks of subject 4: '))
sub5 = float(input('Enter marks of subject 5: '))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
print(f'\nTotal Marks of 5 Subjects: {total_marks}')

marks = total_marks / 5
print(f'\nAverage Marks: {marks}')
print('-'*20)

if marks >= 80 and marks <= 100:
    print('Grade A')
elif marks >= 60 and marks < 80:
    print('Grade B')
elif marks >= 50 and marks < 60:
    print('Grade C')
elif marks >= 40 and marks < 50:
    print('Grade D')
elif marks >= 0 and marks < 40:
    print('Failed')
else:
    print('Invalid Marks. Marks must be in range of 0 to 100.')
```

### Write a program to find a number is odd or even.

```python
number = int(input('Enter a number: '))

if number % 2 == 0:
    print('Even')
else:
    print('Odd')
```

## ATM Amount Withdrawl

```python
amount = 50000
user_pin = 1234

pin = int(input('Enter your pin number: '))
balance = int(input('Enter balance amount for withdrawl: '))

if pin == user_pin and balance < amount:
    print(f'Rs.{balance} withdrawn from A/C 12345678.')
elif pin == user_pin and balance > amount:
    print(f'Insufficient Balance.')
elif pin != user_pin and balance < amount:
    print(f'Incorrect Pin.')
else:
    print('Incorrect Pin Number or Insufficient Balance Amount.')
```
