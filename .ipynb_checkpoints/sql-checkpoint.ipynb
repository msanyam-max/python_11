```python
import sqlite3
```

```python
conn = sqlite3.connect('SQL_file1.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('''
    select * from customers;
''')
data = cursor.fetchall()
print(data)
```

```python
try:
    cursor.execute('''
        Create table customers(
          customer_id integer primary key autoincrement, 
          first_name text not null,
          last_name text not null,
          phone text not null,
          email text not null,
          street text, 
          city text not null
        );
    ''')
    print('Customers table created.')
except:
    print(f'Table customers already exists.')
```

```python
try:
    cursor.execute('''
        Create table staff(
            staff_id integer primary key autoincrement,
            first_name text not null, 
            last_name text not null, 
            phone text not null,
            email text not null
        )
    ''')
    print('Staff table created.')
except:
    print(f'Staff table already exists.')
```

```python
# Insert values in Customer table
```

```python
try:
    cursor.execute('''
        Insert into customers (first_name, last_name, phone, email, street, city)
        values
        ('Harish', 'Khadka', '87654', 'harish@gmail.com', 'Damak', 'Jhapa'),
        ('Rohit', 'Poudel', '54321', 'rohit@gmail.com', 'Hetauda', 'Makwanpur'),
        ('Anish', 'Pandey', '12345', 'anish@gmail.com', 'Prithvi Chowk', 'Pokhara');
        ''')
    conn.commit()
    print('Data inserted successfully.')
except NameError as e:
    print(e)
except:
    print('Data insertion failed.')
```

```python
# Create table
```

```python
try:
    cursor.execute('''
        create table orders(
    	order_id integer primary key autoincrement,
    	order_status text not null,
    	order_date date,
    	customer_id integer,
    	staff_id integer,
    	foreign key (customer_id) references customers(customer_id)
    	foreign key (staff_id) references staff(staff_id)
        );
    ''')
    print('Order table created.')
except:
    print(f'Order table already exists.')
```

```python
# Insert data into table
```

```python
try:
    cursor.execute('''
        Insert into staff (first_name, last_name, phone, email)
        values
        ('David', 'Warner', '911028256', 'warner@gmail.com'),
        ('Shane', 'Warne', '984265789', 'warnie@gmail.com'),
        ('Steve', 'Smith', '9787456123', 'steve@gmail.com');
        ''')
    conn.commit()
    print('Data inserted successfully.')
except:
    print('Data inserton failed.')
```

# Update Table

```python
try:
    cursor.execute('''
        Update customers set first_name = 'Sam', last_name = 'Smith' where customer_id = 1;
        ''')
    conn.commit()
    print('Data inserted successfully.')
except:
    print('Data inserton failed.')
```

# Delete Table

```python
try:
    cursor.execute('''
    delete from orders where order_id = 8;
    ''')
    conn.commit()
    print('Data deleted successfully.')
except:
    print(f'Data deletion failed.')
```
