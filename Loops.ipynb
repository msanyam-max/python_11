```python
# Multiplication Table
n = int(input('Enter number: '))
stop_n = int(input('Enter number: '))
for i in range(1, stop_n+1):
    mul = n * i
    print(f'{n} x {i} = {mul}')
```

```python
remaining = 3
for i in range(1, 3+n):
    remaining = remaining - 1
    print(f'Remaining attempts: {remaining}')
    order_num = int(input('\nEnter your order number: '))

    match order_num:
        case 1:
            print('You have orderd Snaks.')
        case 2:
            print('You have orderd Lunch.')
        case 3:
            print('You have orderd Breakfast.')
        case 4:
            print('You have orderd Dinner.')
        case 5:
            print('Thank you for choosing us.')
            break
        case _:
            print('Invalid Order Number.')
    if order_num > 5:
        remaining = remaining + 1
```

## While Loop

```python
# Find Secret Number
# 1 - 1000
```

```python
# Loop runs until condition becomes false.
secret_number = 7
choose_number = 0

while choose_number != secret_number: # 0 == 7 -> False
    choose_number = int(input('Choose number from 1 - 1000: '))

    if choose_number == secret_number:
        print('Correct Number!!')
    else:
        print('Try Again!!')
```

```python
pin = 1234
amount = 50000
user_pin = 0
attempts = 3

while user_pin != pin:
    attempts -= 1
    print(f'Remaining Attempts: {attempts}')
    
    user_pin = int(input('Enter your pin number: '))
    if user_pin == pin:
        balance = int(input('Enter withdraw amount: '))
        if balance < amount:
            if balance % 1000 == 0:
                print(f'Rs.{balance} is withdrawn from A/C 1234567890.')
                rem_bal = amount - balance
                print(f'Remaining Balance: Rs.{rem_bal}')
            else:
                print('Amount must be multiple of 1000.')
        else:
            print('Balance amount must be less than actual amount.')
    else:
        print('Pin you have entered is incorret.')

    if attempts == 0:
        break
```
