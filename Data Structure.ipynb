```python
# enumerate
'''
- It is used to loop both index and values of a list.
- It always returns 2 values (index and value)
'''

'''
Indexing always starts from 0.
Negative Indexing always starts from -1.
'''
```

```python
names = ['Bob', 'Charles', 'Stuart', 'Edward']
for name in names:
    print(name)
```

```python
for idx, val in enumerate(names):
    print(f'{idx}: {val}')
```

## Sets

```python
# {}
set_a = {1, 2, 3, 4}
type(set_a)
```

```python
a = [3,3,3,4,4,4,3,3,2,1,2,2,1,1,7,6,5,8,9,0,8,7]
set_data = set(a)
print(set_data)
```

```python
type(set_data)
```

```python
set_data.add(1000)
set_data
```

```python
set_data.update([200, 100, 0.3, 0.2, 0.5, 8.5, 300, 3000])
set_data
```

```python
'''
1. remove(value)
2. discard(value)
3. pop()
'''
```

```python
set_data.remove(0.3)
set_data
```

```python
set_data.discard(0.3)
set_data
```

```python
set_data
```
